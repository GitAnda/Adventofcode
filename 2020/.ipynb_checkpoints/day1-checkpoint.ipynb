```python
f = open("day1.txt", "r")
input_puzzle = f.read()
input_puzzle
```

```python
input_puzzle = input_puzzle.split('\n')[0:-1]
input_puzzle = [int(x) for x in input_puzzle]
input_puzzle
```

```python
len(input_puzzle)
```

```python
for i in range(len(input_puzzle)):
    for j in range(i+1, len(input_puzzle)):
        if input_puzzle[i]+input_puzzle[j] == 2020:
            print(input_puzzle[i],input_puzzle[j],input_puzzle[i]*input_puzzle[j])
```

```python
for i in range(len(input_puzzle)):
    for j in range(i+1, len(input_puzzle)):
        for k in range(j+1, len(input_puzzle)):
            if input_puzzle[i]+input_puzzle[j]+input_puzzle[k] == 2020:
                print(input_puzzle[i],input_puzzle[j],input_puzzle[k],input_puzzle[i]*input_puzzle[j]*input_puzzle[k])
```
